BOOK STORE INVENTORY DATA CLEANING:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Book_Store_Inventory.xlsx")  
df
```

```python
print("Original shape:", df.shape)
```

```python
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
corrections = {
    'Title': {
        "The Hitchhiker's Giude to the Galaxy": "The Hitchhiker's Guide to the Galaxy",
        "Prite and Prejudice": "Pride and Prejudice",
        "Sapiens: A brief History of HumanKind": "Sapiens: A Brief History of Humankind",
        "The Zee": "The Book Thief",
        "The Road,Cormac McCarthy": "The Road",
        pd.NA: "The Alchemist"
    },
    'Author': {
        pd.NA: {
            "The Catcher in the Rye": "J.D. Salinger",
            "Atomic Habits": "James Clear"
        }
    },
    'Genre': {
        "None-Fiction": "Non-Fiction",
        "Fiction,": "Fiction"
    },
    'Publication Year': {
        0.202: 2020,
        pd.NA: {
            "The Lean Startup": 2011,
            "The Five People You Meet in Heaven": 2003
        }
    }
} 

df
```

```python
df['Title'] = df['Title'].replace(corrections['Title'])
print(df.head(10))
```

```python
print(df.tail(10))
```

```python
df['Author'] = df.apply(
    lambda row: corrections['Author'][pd.NA][row['Title']] if pd.isna(row['Author']) and row['Title'] in corrections['Author'][pd.NA] else row['Author'],
    axis=1
)

df
```

```python
df['Genre'] = df['Genre'].replace(corrections['Genre'])
df.head(10)
```

```python
df['Publication Year'] = df.apply(
    lambda row: corrections['Publication Year'][pd.NA][row['Title']] if pd.isna(row['Publication Year']) and row['Title'] in corrections['Publication Year'][pd.NA] else row['Publication Year'],
    axis=1
)

df.head(10)
```

```python
df['Publication Year'] = df['Publication Year'].replace({0.202: 2020})
df
```

```python
df['Publication Year'] = df['Publication Year'].astype(int)
```

```python
df['Stock Quantity'].fillna(df['Stock Quantity'].median(), inplace=True)
df.head(10)
```

```python
df.to_excel("cleaned_books_dataset.xlsx", index=False)
print("Cleaned dataset saved as 'cleaned_books_dataset.xlsx'")

df
```

STORING THE CLEANED DATASET IN AN EXCEL DATABASE USING pandas.ExcelWriter

```python
with pd.ExcelWriter("Bookstore_Database.xlsx", engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Books', index=False)

    genre_summary = df['Genre'].value_counts().reset_index()
    genre_summary.columns = ['Genre', 'Count']
    genre_summary.to_excel(writer, sheet_name='Genre Summary', index=False)

    stock_summary = df[['Genre', 'Stock Quantity']].groupby('Genre').sum().reset_index()
    stock_summary.to_excel(writer, sheet_name='Stock Summary', index=False)
```

## DATA VISUALIZATION

```python
#books by genre
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Genre', order=df['Genre'].value_counts().index, palette='viridis')
plt.title("Number of Books by Genre")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
#average bookprice by genre
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Price', y='Genre', estimator='mean', ci=None, palette='mako')
plt.title("Average Book Price by Genre")
plt.xlabel("Average Price")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
#books published over time
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='Publication Year', bins=20, kde=True, color='teal')
plt.title("Distribution of Publication Years")
plt.xlabel("Year")
plt.ylabel("Number of Books")
plt.tight_layout()
plt.show()
```

```python
#top 10 authors by number of books
top_authors = df['Author'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette='cubehelix')
plt.title("Top 10 Authors by Number of Books")
plt.xlabel("Number of Books")
plt.ylabel("Author")
plt.tight_layout()
plt.show()
```
